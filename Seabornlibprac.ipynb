# HeatMap

```python
var=np.linspace(1,10,20).reshape(4,5)
sns.heatmap(data=var,vmin=0,vmax=20,cmap="PuOr",annot=True)#this heatmap show a matrix of 4,5 in color format 
plt.show()
```

```python
var
```

```python
var1=np.linspace(0,10,10).reshape(2,5)
ar=np.array([['a0','a1','a2','a3','a4'],
            ['b0','b1','b2','b3','b4']])
y={'fontsize':20,'color':'r'}
a=sns.heatmap(data=var1,vmin=0,vmax=10,cmap='ocean',annot=ar,fmt='s',annot_kws=y,linewidth=1,linecolor='k',cbar=False,xticklabels=False,yticklabels=False)
a.set(xlabel='columns',ylabel='rows')
sns.set(font_scale=1)
plt.show()
```

# Violin Plot

```python
df=sns.load_dataset('tips')
df
```

```python
sns.violinplot(x='day',y='tip',data=df,hue='time',linewidth=2)
              
a=np.arange(0,11)
plt.yticks(ticks=a)
plt.show()
```

```python
df[df['day']=='Thur'].quantile(0.25)['tip']
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],saturation=0.75,inner='quart')
b=np.arange(0,60,5)
plt.yticks(ticks=b)
plt.show()
```

```python
df[df['time']=='Dinner'].quantile(0.5)['total_bill']
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True,scale='width')
plt.show()
```

```python
sns.countplot(x='day',data=df,hue='sex')
```

```python
sns.violinplot(y='day',x='total_bill',data=df,hue='sex',split=True,scale='width')
plt.show()
```

```python
sns.violinplot(x=df['total_bill'])
plt.show()
```
