```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,40,1000)
d_r=2.5*a
d_s=3*(a-5)
plt.plot(a,d_r,label="distance travelled by thief in time")
plt.plot(a,d_s,label="Distance travelled by sheriff in time")
plt.title("Distance time graph of thief and sheriff")
plt.xlabel("time taken")
plt.ylabel("Distance Travelled")
plt.axhline(y=75,linestyle='--')
plt.axvline(x=30,linestyle='--')
plt.legend()
plt.show()
```
