```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df
```

```python
sns.kdeplot(x="sepal_length",data=df)
```

```python
temp=sorted(df["sepal_length"].tolist())
```

```python
x_quants=[]
for i in range(1,101):
    x_quants.append(np.percentile(temp,i))
```

```python
y=np.random.normal(loc=0,scale=1,size=1000)
```

```python
y_quants=[]
for i in range(1,101):
    y_quants.append(np.percentile(y,i))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_quants,y_quants)
```

```python
import statsmodels.api as sm
fig=sm.qqplot(df['sepal_length'],line='45',fit=True)
plt.show()
```

Uniform Plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm

x=np.random.uniform(low=0,high=1,size=1000)
```

```python
plt.hist(x)
```

```python
dist=stats.uniform.fit(x)
params=stats.uniform(loc=dist[0],scale=dist[1])
```

```python
fig=sm.qqplot(x,dist=params,line='45')
plt.show()
```
