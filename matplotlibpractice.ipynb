```python
import csv
from collections import Counter
with open("data.csv") as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
language=[]
popularity=[]
for i in language_counter.most_common(15):
    language.append(i[0])
    popularity.append(i[1])
language.reverse()
popularity.reverse()
plt.style.use('fivethirtyeight')
plt.barh(language,popularity,color='r')
plt.title("Popularity in diffent coding languages")
plt.ylabel("Diffient Coding Languages")
plt.xlabel("Popularity")
# print(plt.style.available)
plt.show()
```

```python
#language popularity
from collections import Counter
a=pd.read_csv("data.csv",delimiter=',')
b=a['LanguagesWorkedWith']
lang_counter=Counter()
for i in b:
    lang_counter.update(i.split(';'))
slices=[]
labels=[]
for i in lang_counter.most_common(15):
    slices.append(i[1])
    labels.append(i[0])
plt.pie(slices,labels=labels,autopct='%.2f %%')
plt.show()
    
```

```python
from datetime import datetime
from matplotlib import dates as mpl_dates
dates=[
    datetime(2019,5,24),
    datetime(2019,5,25),
    datetime(2019,5,26),
    datetime(2019,5,27),
    datetime(2019,5,28),
    datetime(2019,5,29),
    datetime(2019,5,30)
]
y=[0,1,3,4,6,7,9]
plt.figure(figsize=(10,5))
plt.plot_date(dates,y,linestyle='solid') #marker=None
plt.gcf().autofmt_xdate() #gcf=get configure tilt the date
date_format=mpl_dates.DateFormatter("%b,%d %Y")
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```
